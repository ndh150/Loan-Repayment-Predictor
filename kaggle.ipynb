```python
def clear_percent (row):
    try:
        d = float(row['interest_rate'][:-1])/100.
    except:
        d = None
    return d

def clear_percent_debt (row):
    try:
        d = float(row['debttoincomeratio'][:-1])/100.
    except:
        d = None
    return d

def clear_percent_revolving (row):
    try:
        d = float(row['revolving_line_utilization'][:-1])/100.
    except:
        d = None
    return d

loans['int_rate_clean'] = loans.apply (lambda row: clear_percent(row),axis=1)
loans['debttoincomeratio_clean'] = loans.apply (lambda row: clear_percent_debt(row),axis=1)
loans['revolving_line_utilization_clean'] = loans.apply (lambda row: clear_percent_revolving(row),axis=1)
loans.head()

loans_predict['int_rate_clean'] = loans_predict.apply (lambda row: clear_percent(row),axis=1)
loans_predict['debttoincomeratio_clean'] = loans_predict.apply (lambda row: clear_percent_debt(row),axis=1)
loans_predict['revolving_line_utilization_clean'] = loans_predict.apply (lambda row: clear_percent_revolving(row),axis=1)
loans_predict.head()
```

```python
def clear_term (row):
    try:
        if row['loan_length']=='36 months':
            d = 1
        else:
            if row['loan_length']=='60 months':
                d = 2
            else:
                if np.isnan(row['loan_length']):
                    d = None
                else:
                    print ('WRONG')
                    print (row['loan_length'])
    except:
        print ('EXCEPT')
        d = None
    return d

loans['loan_length_clean'] = loans.apply (lambda row: clear_term(row),axis=1)
loans_predict['loan_length_clean'] = loans_predict.apply (lambda row: clear_term(row),axis=1)

loans.head()

def clean_emp_length(argument):
    switcher = {
        '1 year': 1,
        '2 years': 2,
        '3 years': 3,
        '4 years': 4,
        '5 years': 5,
        '6 years': 6,
        '7 years': 7,
        '8 years': 8,
        '9 years': 9,
        '10+ years': 10,
        '< 1 year': 0,
        'n/a':None,
    }
    try:
        d = switcher[argument['employment_length']]    
    except:
        d = None
    return d

loans['employment_length_clean'] = loans.apply (lambda row: clean_emp_length(row),axis=1)
loans_predict['employment_length_clean'] = loans_predict.apply (lambda row: clean_emp_length(row),axis=1)

loans.head()
#np.unique(loans['loan_length'])
np.unique(loans['loan_purpose'])
```

```python
import xgboost as xgb
from sklearn.metrics import precision_score

dtrain = xgb.DMatrix(X, label=y)
dtest = xgb.DMatrix(X_test)

param = {
    'max_depth': 4,  # the maximum depth of each tree
    'eta': 0.3,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 44
}  # the number of classes that exist in this datset
num_round = 20  # the number of training iterations

bst = xgb.train(param, dtrain, num_round)
preds = bst.predict(dtest)

best_preds = np.asarray([np.argmax(line) for line in preds])
X_test['preds']= best_preds
X_test.head()

submission = X_test[['loan_id','preds']]
#print (precision_score(y_test, best_preds, average='macro'))
submission.to_csv('SubmissionDayFinal.csv', encoding='utf-8', index=False)
submission.head()
```
